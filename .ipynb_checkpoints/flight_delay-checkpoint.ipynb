## Import Dataset

```python
import numpy as np
import pandas as pd
import pickle
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv('flightdata.csv')
df.head()
```

```python
df['DEST'].unique()
```

## Dataset Cleaning 

```python
df.info()
```

```python
df=df.drop(['Unnamed: 25'],axis=1)
```

```python
fig, ax =plt.subplots(figsize=(15,10))
sns.heatmap(df.iloc[:,12:].corr(),annot=True,ax=ax)
plt.show()
```

```python
df.describe()
```

```python
df=df[['FL_NUM','MONTH','DAY_OF_MONTH','DAY_OF_WEEK','ORIGIN','DEST','DEP_DEL15','CRS_ARR_TIME','ARR_DEL15']]
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(df['DEP_DEL15'].mode()[0], inplace=True)
df.fillna(df['ARR_DEL15'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
fig,ax=plt.subplots(figsize=(5,6))
sns.boxplot(data=df['CRS_ARR_TIME'])
plt.show()
```

```python
df=pd.get_dummies(df,columns=['ORIGIN','DEST']) #performed one-hot encoding
df.head()
```

## Splitting data features and target variable

```python
X=df.drop(columns=['ARR_DEL15'])
X
```

```python
X.columns
```

```python
y=df[['ARR_DEL15']]
y
```

## Splitting Testing and Training Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
y_train.value_counts()
```

## Appling ML Models

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
decision_tree_classifier=DecisionTreeClassifier()
decision_tree_classifier=decision_tree_classifier.fit(X_train,y_train)
```

```python
decision_tree_prediction=decision_tree_classifier.predict(X_test)
```

### performance metrics

```python
print("Confusion Metrics : - \n",confusion_matrix(decision_tree_prediction,y_test))
print("\nAccuracy Score : - ",accuracy_score(decision_tree_prediction,y_test))
```

### SVC ;- support vector classifier

```python
from sklearn.svm import SVC
SVC_model = SVC()
SVC_model.fit(X_train,y_train)
SVC_prediction=SVC_model.predict(X_test)
```

### performance metrics

```python
print("Confusion Metrics : - \n",confusion_matrix(SVC_prediction,y_test))
print("\nAccuracy Score : - ",accuracy_score(SVC_prediction,y_test))
```

### KNN - K-nearest neighbour

```python
from sklearn.neighbors import KNeighborsClassifier
KNN_model=KNeighborsClassifier(n_neighbors=5)
KNN_model.fit(X_train,y_train)
KNN_prediction=KNN_model.predict(X_test)
```

### performance metrics

```python
print("Confusion Metrics : - \n",confusion_matrix(KNN_prediction,y_test))
print("\nAccuracy Score : - ",accuracy_score(KNN_prediction,y_test))
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR_model = LogisticRegression(random_state=0)
LR_model.fit(X_train, y_train)
LR_prediction = LR_model.predict(X_test)
```

### performance metrics

```python
print("Confusion Metrics : - \n",confusion_matrix(LR_prediction,y_test))
print("\nAccuracy Score : - ",accuracy_score(LR_prediction,y_test))
```

### ensemble model for best three

```python
from sklearn.ensemble import VotingClassifier
ensemble=VotingClassifier(estimators=[('dt',decision_tree_classifier),('svc',SVC_model),('lr',LR_model)],voting='hard')
ensemble.fit(X_train,y_train)
ensemble_prediction=ensemble.predict(X_test)
```

### prediction metrics

```python
print("Confusion Metrics : - \n",confusion_matrix(ensemble_prediction,y_test))
print("\nAccuracy Score : - ",accuracy_score(ensemble_prediction,y_test))
```

```python
joblib.dump(LR_model,'flight.pkl')
```
